```python
import numpy as np
import pandas as pd
```

```python
dados = pd.read_csv('funcionarios_dados_ti.csv', sep = ';')
dados
```

```python
salario_array = np.array(dados['Salario'])
```

```python
media = np.mean(salario_array)
mediana = np.median(salario_array)
distancia = (media - mediana) / mediana

print(f'Média: {media}')
print(f'Mediana: {mediana}')
print(f'Distancia: {round(distancia*100,2)}%') 
print('Nesse caso, a média é confiavel!')
```

```python
# Na última década, algum município da região metropolitana pode ser considerado outlier sobre as ocorrências de furto de veículos?

municipio = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseMunicipioMensal.csv', sep = ';',encoding='latin1')
municipio
```

```python
furto_veiculos = municipio.groupby('fmun')['furto_veiculos'].sum().reset_index()
furto_veiculos
```

```python
furto_veiculos_array = np.array(furto_veiculos['furto_veiculos'])
```

```python
mediana = np.median(furto_veiculos_array)
media = np.mean(furto_veiculos_array)

q1 = np.percentile(furto_veiculos_array, 25)
q2 = np.percentile(furto_veiculos_array, 50)
q3 = np.percentile(furto_veiculos_array, 75)
distancia = (media - mediana) / mediana 
iqr = q3 - q1
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)

print(f'Média: {media:.2f}')
print(f'Mediana: {mediana}')
print(f'Distancia: {distancia * 100:.2f}%')
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2, Mediana): {q2}')
print(f'Terceiro quartil geral (Q3): {q3}')
```

```python
furtos_municipio = furto_veiculos.loc[furto_veiculos['furto_veiculos'] >= limite_superior]
furtos_municipio.sort_values(by = 'furto_veiculos', ascending= False)

```
