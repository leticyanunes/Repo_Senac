# Realizar um estudo sobre os preços praticados nos bairros da cidade do rio de janeiro. sua tarefa é analisar e identificar os bairros com os seguintes criterios: bairros com o menor valor médio de venda e com maior valor medio de venda. etanol, gasolina e diesel 

```python
import pandas as pd # Biblioteca Python para análise e manipulaçãode dados, trabalha com tabelas (DataFrames) e séries (Series)
```

```python

df = pd.read_csv('Combustivel.csv',sep=';',encoding='latin1')

df.head() #olha as primeiras linhas para entender a tabela
```

```python
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
df.columns #listar as colunas
```

```python
df_rj = df.loc[df['Estado - Sigla'] == 'RJ']
df_rj
```

```python
etanol = df_rj.loc[df['Produto'] == 'ETANOL']
etanol
```

```python
gasolina = df_rj.loc[df['Produto'] == 'GASOLINA']
gasolina
```

```python
diesel = df_rj.loc[df['Produto'] == 'DIESEL']
diesel
```

```python
media_etanol = etanol.groupby('Bairro')['Valor de Venda'].mean().reset_index()
media_etanol
```

```python
media_gasolina = gasolina.groupby('Bairro')['Valor de Venda'].mean().reset_index()
media_gasolina
```

```python
media_diesel = diesel.groupby('Bairro')['Valor de Venda'].mean().reset_index()
media_diesel
```
