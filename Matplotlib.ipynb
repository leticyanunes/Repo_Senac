```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'Ano': [2020, 2021, 2022, 2023], 'Vendas': [100, 150, 130, 180]})

df['Ano'] = df['Ano'].astype(str) #transformar ano em string

plt.plot(df['Ano'], df['Vendas']) # 1° X , 2° Y
plt.title('Vendas por Ano')
plt.xlabel('Ano')
plt.ylabel('Vendas')
plt.grid(True)
plt.show()
```

```python
df = pd.DataFrame({'Produto': ['A', 'B', 'C'], 'Vendas': [120, 90, 150]})

df.sort_values(by = 'Vendas', ascending = False, inplace = True) #inplace = True é pra n precisar colocar o nome do Dataframe na frente. ex: df = ....

plt.bar(df['Produto'], df['Vendas'])
plt.title('Vendas por Produto')
plt.xlabel('Produtos')
plt.ylabel('Vendas')
plt.show()
```

```python
dados = {'Ano': [2018, 2019, 2020, 2021, 2022], 'Vendas': [150, 200, 250, 300, 350]}

df = pd.DataFrame(dados)

plt.figure(figsize = (10,7)) #determina o tamanho da imagem
plt.barh(df['Ano'], df['Vendas'], color = 'purple')
plt.title('Vendas por Ano')
plt.xlabel('Vendas')
plt.ylabel('Ano')
plt.savefig('grafico barras horizontal.png') # comando para salvar o grafico
plt.show()
```

```python
df = {'dias': ['Segunda', 'Terça', 'Quarta', 'Quinta', 'Sexta'], 'Venda': [500, 200, 1000, 150, 1950]}

dias = df['dias']
valores = df['Venda']

plt.plot(dias, valores, marker = 'o', linestyle = '-', color = 'green') # marker = pra marcar os valores, a bolinha / linestyle = tipo de linha
plt.title('Vendas por Dias')
plt.xlabel('Dia da Semana')
plt.ylabel('Valor Vendido')
plt.show()
```
