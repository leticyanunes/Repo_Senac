```python
import pandas as pd
```

```python
caramelo = pd.read_csv('CARAMELO (1).csv', sep=';')
caramelo
```

```python
caramelo['Valor de Venda'] = caramelo['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
# qual cidade lucra mais *
# qual sexo e a idade dos clientes * 
# qual a marca favorita dos clientes *
# qual categoria e produto mais vendido [qual mais lucrativo(nem sempre o mais vendido é o mais lucrativo)] *
# qual forma de pagamento mais usada *
```

```python
caramelo.columns
```

```python
lucro_cidade = caramelo.groupby('Cidade Loja')['Valor de Venda'].sum().reset_index().sort_values(by='Valor de Venda', ascending=False) # ascending = false (maior para o menor), true = menor para o maior
lucro_cidade #CIDADE MAIS LUCRATIVA
```

```python
clientes = caramelo.groupby('Sexo')['Idade'].median().reset_index() #calcula a mediana
clientes = clientes.rename(columns={'Idade': 'Média de Idade'})
clientes #mediana de idade por genero
```

```python
marca = caramelo.groupby('Marca')['Nome Cliente'].count().reset_index() #count é para n concatenar e sim contar
marca = marca.rename(columns={'Nome Cliente': 'Qtd_Clientes'}) #mudar o nome da coluna
marca # Qual a marca favorita 
```

```python
pagamento = caramelo.groupby('Metodo de Pagamento')['ID Cliente'].count().reset_index() 
pagamento = pagamento.rename(columns={'ID Cliente': 'Qnt_Clientes'})
pagamento # metodo de pagamento mais utilizado 
```

```python
produto = caramelo.groupby('Categoria Produto')['Produto'].count().reset_index()
produto #categoria mais vendida
```

```python
produto_lucrativo = caramelo.groupby('Produto')['Valor de Venda'].sum().reset_index().sort_values(by='Valor de Venda', ascending=False)
produto_lucrativo #produto que mais gera receita
```
