```python
import numpy as np
import pandas as pd
```

```python
projeto = pd.read_csv('https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv', sep = ';',encoding='latin1')
projeto
```

```python
baixada = projeto[projeto["regiao"] == "Baixada Fluminense"]
baixada
```

```python
menores = baixada.groupby('munic')['aaapai'].sum().reset_index()
menores
```

```python
apreensao_menores_array = np.array(menores['aaapai'])
```

```python
mediana = np.median(apreensao_menores_array)
media = np.mean(apreensao_menores_array)

q1 = np.percentile(apreensao_menores_array, 25)
q2 = np.percentile(apreensao_menores_array, 50)
q3 = np.percentile(apreensao_menores_array, 75)
distancia = (media - mediana) / mediana

iqr = q3 - q1 
limite_superior = q3 + (1.5 * iqr)
limite_inferior = q1 - (1.5 * iqr)

print(f'Média: {media:.2f}')
print(f'Mediana: {mediana}')
print(f'Distancia: {distancia * 100:.2f}%')
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2, Mediana): {q2}')
print(f'Terceiro quartil geral (Q3): {q3}')
```

```python
total_apreensao = menores.loc[menores['aaapai'] >= limite_superior]
total_apreensao.sort_values(by = 'aaapai', ascending = False)
```

```python
total_apreensao.to_csv('outliers_baixada', index = False)
```

```python
menores.to_csv('munic_baixada', index = False)
```
